```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Downloads/Global_terrorist.csv', encoding="ISO-8859-1")
```

```python
df
```

```python
df['region_txt'].unique()
```

```python

import folium
from folium.plugins import HeatMap

filter1 = df[df['region_txt'] == 'Middle East & North Africa']

filter2 = filter1[filter1['attacktype1_txt'] == 'Bombing/Explosion']

filter3 = filter2[filter2['iyear'] >= 2000]

filter4 = filter3[filter3['nkill'] >= 10]

Mer = filter4.groupby(['longitude', 'latitude']).size().reset_index(name='incident_count')

heat_df = Mer[['latitude', 'longitude', 'incident_count']].values.tolist()

my_map = folium.Map(location=[30, 40], zoom_start=4)
HeatMap(heat_df).add_to(my_map)

my_map
```
